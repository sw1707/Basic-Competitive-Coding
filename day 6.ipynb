```python
# 1) Search target; if not found, return nearest greater element, else -1.
def search_or_next_greater(arr, target):
    found = False
    nearest_greater = None

    for value in arr:
        if value == target:
            found = True
            break
        if value > target:
            if nearest_greater is None or value < nearest_greater:
                nearest_greater = value

    if found:
        return target
    if nearest_greater is not None:
        return nearest_greater
    return -1

print(search_or_next_greater([10, 25, 7, 40], 20))
```

```python
# 2) Find first and last index of target using linear search only.
def first_last_index(arr, target):
    first = -1
    last = -1

    i = 0
    while i < len(arr):
        if arr[i] == target:
            if first == -1:
                first = i
            last = i
        i = i + 1

    return first, last

print(first_last_index([4, 2, 7, 2, 9, 2], 2))
```

```python
# 3) Count target occurrences and return all positions.
def count_and_positions(arr, target):
    count = 0
    positions = []

    i = 0
    while i < len(arr):
        if arr[i] == target:
            count = count + 1
            positions.append(i)
        i = i + 1

    return count, positions

print(count_and_positions([1, 3, 3, 5, 3], 3))
```

```python
# 4) Find index of element closest to target (minimum absolute difference).
def closest_index(arr, target):
    if len(arr) == 0:
        return -1

    best_index = 0
    best_diff = abs(arr[0] - target)

    i = 1
    while i < len(arr):
        current_diff = abs(arr[i] - target)
        if current_diff < best_diff:
            best_diff = current_diff
            best_index = i
        i = i + 1

    return best_index

print(closest_index([10, 22, 35, 40], 28))
```

```python
# 5) Search mixed-type list and return indices with exact type + value match.
def mixed_type_search(arr, target):
    indices = []

    i = 0
    while i < len(arr):
        if arr[i] == target and type(arr[i]) == type(target):
            indices.append(i)
        i = i + 1

    return indices

print(mixed_type_search([10, 10.0, "10", 10], 10))
```

```python
# 6) Linear search in a list of lists. Return (row, column) if found.
def search_in_matrix(matrix, target):
    row = 0
    while row < len(matrix):
        col = 0
        while col < len(matrix[row]):
            if matrix[row][col] == target:
                return (row, col)
            col = col + 1
        row = row + 1
    return (-1, -1)

print(search_in_matrix([[1, 2], [3, 4], [5, 6]], 4))
```

```python
# 7) Perform linear search from both ends simultaneously and return index which finds it first.
def both_ends_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        if arr[left] == target:
            return left
        if arr[right] == target:
            return right
        left = left + 1
        right = right - 1

    return -1

print(both_ends_search([9, 3, 7, 1, 7, 5], 7))
```

```python
# 8) Find first element greater than target and first element smaller than target.
def first_greater_and_smaller(arr, target):
    first_greater = -1
    first_smaller = -1

    i = 0
    while i < len(arr):
        if first_greater == -1 and arr[i] > target:
            first_greater = arr[i]
        if first_smaller == -1 and arr[i] < target:
            first_smaller = arr[i]
        if first_greater != -1 and first_smaller != -1:
            break
        i = i + 1

    return first_greater, first_smaller

print(first_greater_and_smaller([12, 8, 15, 5, 20], 10))
```

```python
# 9) Search for a substring inside each string of a list of strings. Return all indices where substring appears.
def substring_indices(words, sub):
    indices = []
    i = 0
    while i < len(words):
        if sub in words[i]:
            indices.append(i)
        i = i + 1
    return indices

print(substring_indices(["apple", "banana", "grape", "pineapple"], "app"))
```

```python
# 10) Perform linear search but skip every k-th index (skip index 2, 5, 8, ... for k=3).
def search_skip_kth(arr, target, k):
    i = 0
    while i < len(arr):
        if (i + 1) % k == 0:
            i = i + 1
            continue
        if arr[i] == target:
            return i
        i = i + 1
    return -1

print(search_skip_kth([11, 22, 33, 44, 55, 66, 77], 55, 3))
```

```python
# 11) Search in a rotated list using linear search.
def search_rotated_linear(arr, target):
    i = 0
    while i < len(arr):
        if arr[i] == target:
            return i
        i = i + 1
    return -1

print(search_rotated_linear([30, 40, 50, 10, 20], 10))
```

```python
# 12) Search for target and return how many comparisons were made.
def search_with_comparisons(arr, target):
    comparisons = 0
    i = 0
    while i < len(arr):
        comparisons = comparisons + 1
        if arr[i] == target:
            return i, comparisons
        i = i + 1
    return -1, comparisons

print(search_with_comparisons([5, 9, 12, 18], 18))
```

```python
# 13) Find the index of second occurrence of target. If not found -> print "Not enough occurrences".
def second_occurrence_index(arr, target):
    count = 0
    i = 0
    while i < len(arr):
        if arr[i] == target:
            count = count + 1
            if count == 2:
                return i
        i = i + 1
    return "Not enough occurrences"

print(second_occurrence_index([2, 4, 2, 6, 2], 2))
```

```python
# 14) Search for multiple targets in the same list. Return dictionary: target -> indices.
def multi_target_search(arr, targets):
    result = {}

    t = 0
    while t < len(targets):
        target = targets[t]
        positions = []

        i = 0
        while i < len(arr):
            if arr[i] == target:
                positions.append(i)
            i = i + 1

        result[target] = positions
        t = t + 1

    return result

print(multi_target_search([1, 2, 3, 2, 4, 1], [1, 2, 5]))
```

```python
# 15) Find the longest string in a list using linear search only.
def longest_string(words):
    if len(words) == 0:
        return ""

    longest = words[0]
    i = 1
    while i < len(words):
        if len(words[i]) > len(longest):
            longest = words[i]
        i = i + 1

    return longest

print(longest_string(["cat", "elephant", "dog", "giraffe"]))
```

```python
# 16) Reverse direction linear search (right to left).
def reverse_linear_search(arr, target):
    i = len(arr) - 1
    while i >= 0:
        if arr[i] == target:
            return i
        i = i - 1
    return -1

print(reverse_linear_search([10, 20, 30, 20, 10], 20))
```

```python
# 17) Tolerance search: allow up to +/-2 difference and return index.
def tolerance_search(arr, target):
    i = 0
    while i < len(arr):
        if abs(arr[i] - target) <= 2:
            return i
        i = i + 1
    return -1

print(tolerance_search([40, 47, 53, 60], 50))
```

```python
# 18) Linear search on a circular list from given start index until it wraps back.
def circular_linear_search(arr, target, start_index):
    if len(arr) == 0:
        return -1

    i = start_index
    while True:
        if arr[i] == target:
            return i
        i = i + 1
        if i == len(arr):
            i = 0
        if i == start_index:
            break

    return -1

print(circular_linear_search([7, 9, 2, 5], 5, 2))
```

```python
# 19) Find the element that gives maximum absolute difference from target.
def max_abs_diff_element(arr, target):
    if len(arr) == 0:
        return -1

    max_element = arr[0]
    max_diff = abs(arr[0] - target)

    i = 1
    while i < len(arr):
        current_diff = abs(arr[i] - target)
        if current_diff > max_diff:
            max_diff = current_diff
            max_element = arr[i]
        i = i + 1

    return max_element

print(max_abs_diff_element([3, 15, 28, 50], 20))
```

```python
# 20) Search in list of dictionaries where a key matches target and return index.
def search_dict_by_key(dict_list, key_name, target):
    i = 0
    while i < len(dict_list):
        if key_name in dict_list[i] and dict_list[i][key_name] == target:
            return i
        i = i + 1
    return -1

print(search_dict_by_key([{"id": 101}, {"id": 102}, {"id": 103}], "id", 102))
```
