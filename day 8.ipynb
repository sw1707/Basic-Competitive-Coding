Problem Statement

Description

A company is transmitting data to another server. The data is in the form of
numbers.
To secure the data during transmission, they plan to obtain a security key
that will be sent along with the data.
The security key is identified as the count of repeating digits in the data.

Write an algorithm to find the security key for the given data.

Input

The input consists of an integer value representing the data to be
transmitted.

Output

Print an integer representing the security key for the given data.
If no digit is repeated, print -1.

Example

Input:
12234

Output:
1

```python
data = input("Enter digit: ")
print("You entered: ", data)
count = 0
for i in range(len(data)):
    for j in range(i + 1, len(data)):
        if data[i] == data[j]:
            count += 1
            break
        
if count == 0:    
    print(-1)
else:    
    print("Security key = ", count)
```

Take set of numbers as input from users and seperate it such that odd numbers comes after even numbers

```python
# Take set of numbers as input from users and seperate it such that odd numbers comes after even numbers

size = int(input("Enter size of list: "))
numbers = []
for i in range(size):
    num = int(input("Enter number: "))
    numbers.append(num)

even = []
odd = []

for num in numbers:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)

result = even + odd
print("Separated list: ", result)
```
