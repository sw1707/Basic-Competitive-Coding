```python
class Node:
    def __init__(self, info):
        self.info = info
        self.link = None

class LinkedList:
    def __init__(self):
        self.start = None

    def insert(self):
        info = int(input("Enter any number: "))
        temp = Node(info)
        if self.start is None:
            self.start = temp
        else:
            t1 = self.start
            while t1.link is not None:
                t1 = t1.link
            t1.link = temp

    def delete_node(self):
        if self.start is None:
            print("List is empty")
        else:
            self.start = self.start.link

    def print_list(self):
        if self.start is None:
            print("List is empty")
        else:
            t2 = self.start
            while t2 is not None:
                print(t2.info, end=" ")
                t2 = t2.link
            print()

    def menu(self):
        print("1.ADD\n2.Delete\n3.Print\n4.Exit\n")
        return int(input("Enter your choice: "))

    def run(self):
        while True:
            choice = self.menu()
            if choice == 1:
                self.insert()
            elif choice == 2:
                self.delete_node()
            elif choice == 3:
                self.print_list()
            elif choice == 4:
                print("Exiting...")
                break
            else:
                print("Invalid entry")

if __name__ == "__main__":
    l1 = LinkedList()
    l1.run()
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def reverseList(self, head):
        previous = None
        currentPOS = head

        while currentPOS is not None:
            nxt = currentPOS.next
            currentPOS.next = previous
            previous = currentPOS
            currentPOS = nxt

        return previous

# demo output
n1 = ListNode(1)
n2 = ListNode(2)
n3 = ListNode(3)
n1.next = n2
n2.next = n3

reversed_head = Solution().reverseList(n1)

curr = reversed_head
while curr is not None:
    print(curr.val, end=" ")
    curr = curr.next
print()
```

```python
class Node:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def hasCycle(self, head):
        slow = head
        fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True
        return False

# demo output
# list with cycle: 1 -> 2 -> 3 -> 2 ...
a = Node(1)
b = Node(2)
c = Node(3)
a.next = b
b.next = c
c.next = b

# list without cycle: 4 -> 5 -> None
x = Node(4)
y = Node(5)
x.next = y

s = Solution()
print("Has cycle (expected True):", s.hasCycle(a))
print("Has cycle (expected False):", s.hasCycle(x))
```

```python
class Node:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def detectCycle(self, head):
        slow = head
        fast = head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                break
        else:
            return None

        slow = head
        while slow != fast:
            slow = slow.next
            fast = fast.next

        return slow

# demo output
# cycle starts at node value 20
p = Node(10)
q = Node(20)
r = Node(30)
s = Node(40)
p.next = q
q.next = r
r.next = s
s.next = q

ans = Solution().detectCycle(p)
if ans is None:
    print("No cycle")
else:
    print("Cycle starts at:", ans.val)
```
