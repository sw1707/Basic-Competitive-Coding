```python
# To take a user's input name and age and print a message

name = input("Enter your name: ")
age = input("Enter your age: ")
print(f"Hello {name}! You are {age} years old.")
```

```python
# WAP to calculate the perimeter of a rectangle given its length and width

length = int(input("Enter the length of the rectangle: "))
width = int(input("Enter the width of the rectangle: "))
perimeter = 2 * (length + width)
print(f"The perimeter of the rectangle is: {perimeter}")
```

```python
# Convert kilometers into meters and centimeters

kilometers = float(input("Enter distance in kilometers: "))
meters = kilometers * 1000
centimeters = meters * 100
print(f"{kilometers} kilometers is equal to {meters} meters and {centimeters} centimeters.")
```

```python
# calculate total bill if a user buys 3 products (take price of each product as input)

price1 = int(input("Enter the price of product 1: "))
price2 = int(input("Enter the price of product 2: "))
price3 = int(input("Enter the price of product 3: "))

total_bill = price1 + price2 + price3

print(f"The total bill for the 3 products is: {total_bill}")
```

```python
# Convert hours into minutes and seconds

hours = int(input("Enter the number of hours: "))
minutes = hours * 60
seconds = minutes * 60
print(f"{hours} hours is equal to {minutes} minutes and {seconds} seconds.")
```

```python
# calculate the area of a triangle given its base and height

base = int(input("Enter the base of the triangle: "))
height = int(input("Enter the height of the triangle: "))
area = (base * height) / 2
print(f"The area of the triangle is: {area}")
```

```python
# take a number and print its double and triple

number = int(input("Enter a number: "))
double = number * 2
triple = number * 3
print(f"The double of {number} is: {double}")
print(f"The triple of {number} is: {triple}")
```

```python
# calculate the total number of seconds in a given no of days

days = int(input("Enter the number of days: "))
total_seconds = days * 24 * 60 * 60
print(f"The total number of seconds in {days} days is: {total_seconds}")
```

```python
# convery rupees into paise

rupees = float(input("Enter the amount in rupees: "))
paise = rupees * 100
print(f"{rupees} rupees is equal to {paise} paise.")
```

```python
# Calculate total number of students in a class by taking the number of boys and girls as input

boys = int(input("Enter the number of boys: "))
girls = int(input("Enter the number of girls: "))
total_students = boys + girls
print(f"The total number of students in the class is: {total_students}")
```
