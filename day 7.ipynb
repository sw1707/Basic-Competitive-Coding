```python
# Linear Search Algorithm
def linear_search(list, target):
    for i in range(len(list)):
        if list[i] == target:
            return i  
    return -1  

list = [10, 20, 30, 40, 50]
target = int(input("Enter the target value to search: "))
index = linear_search(list, target)
print(f"Target found at index: {index}")
```

```python
# 1️⃣ Search an element but return the nearest greater element if not found.
# If target is not present, return the element just greater than target.
# If no greater element exists → return -1.

list = [10, 20, 30, 40, 50]
target = int(input("Enter the target value to search: "))
nearest_greater = -1
for i in range(len(list)):
    if list[i] == target:
        print(f"Target found at index: {i}")
        break
    elif list[i] > target:
        nearest_greater = list[i]
        break 
        
print(f"Target not found. Nearest greater element: {nearest_greater}")
```

```python
# 2️⃣ Find the first and last index of a target in a list using linear search only.
# (Do not use built-in functions.)

def linear_search_first_last(list, target):
    first_index = -1
    last_index = -1
    for i in range(len(list)):
        if list[i] == target:
            if first_index == -1:
                first_index = i  
            last_index = i  
    return first_index, last_index

list = [10, 20, 30, 40, 50, 10]
target = int(input("Enter the target value to search: "))
first_index, last_index = linear_search_first_last(list, target)
print(f"First index: {first_index}, Last index: {last_index}")
```

```python
# Binary Search
def binary_search(list,target):
    low=0
    high=len(list)-1
    while low<=high:
        mid=(low+high)//2
        if  list[mid]==target:
            return mid
        elif list[mid]<target:
            low=mid+1
        else:
            high=mid-1
    return -1    

list=[10,20,30,40,50]
target=int(input("Enter the target value to search: "))
result=binary_search(list,target)
print(f"Target found at index: {result}")
```
