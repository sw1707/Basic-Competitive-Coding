```python
# 1. Write a function to check if a number is even or odd.
def check_even_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(check_even_odd(7))
```

```python
# 2. Write a function to find the largest of three numbers.
def largest_of_three(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print(largest_of_three(10, 25, 18))
```

```python
# 3. Write a function to check whether a number is positive, negative or zero.
def check_number_type(num):
    if num > 0:
        return "Positive"
    elif num < 0:
        return "Negative"
    else:
        return "Zero"

print(check_number_type(-3))
```

```python
# 4. Write a function to calculate the factorial of a number using a while loop.
def factorial_while(n):
    result = 1
    while n > 1:
        result = result * n
        n = n - 1
    return result

print(factorial_while(5))
```

```python
# 5. Write a function to print numbers from 1 to N using a for loop.
def print_1_to_n(n):
    for i in range(1, n + 1):
        print(i)

print_1_to_n(5)
```

```python
# 6. Write a function to count even and odd numbers in a list.
def count_even_odd(numbers):
    even_count = 0
    odd_count = 0
    for num in numbers:
        if num % 2 == 0:
            even_count = even_count + 1
        else:
            odd_count = odd_count + 1
    return even_count, odd_count

print(count_even_odd([1, 2, 3, 4, 5, 6]))
```

```python
# 7. Write a function to reverse a number using while loop.
def reverse_number(num):
    rev = 0
    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num = num // 10
    return rev

print(reverse_number(1234))
```

```python
# 8. Write a function to check if a number is prime or not.
def is_prime(num):
    if num <= 1:
        return False
    i = 2
    while i < num:
        if num % i == 0:
            return False
        i = i + 1
    return True

print(is_prime(11))
```

```python
# 9. Write a function to calculate the sum of digits of a number.
def sum_of_digits(num):
    total = 0
    while num > 0:
        total = total + (num % 10)
        num = num // 10
    return total

print(sum_of_digits(567))
```

```python
# 10. Write a function to print the multiplication table of a number.
def multiplication_table(num):
    for i in range(1, 11):
        print(num, "x", i, "=", num * i)

multiplication_table(5)
```

```python
# 11. Write a function to determine whether a year is a leap year.
def is_leap_year(year):
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        return True
    else:
        return False

print(is_leap_year(2024))
```

```python
# 12. Write a function to find the largest element in a list without using max().
def largest_in_list(values):
    largest = values[0]
    for value in values:
        if value > largest:
            largest = value
    return largest

print(largest_in_list([5, 9, 2, 14, 7]))
```

```python
# 13. Write a function to count vowels in a string.
def count_vowels(text):
    count = 0
    for ch in text:
        if ch == 'a' or ch == 'e' or ch == 'i' or ch == 'o' or ch == 'u' or ch == 'A' or ch == 'E' or ch == 'I' or ch == 'O' or ch == 'U':
            count = count + 1
    return count

print(count_vowels("Education"))
```

```python
# 15. Write a function to find the second largest number in a list.
def second_largest(values):
    first = values[0]
    second = values[0]

    for value in values:
        if value > first:
            second = first
            first = value
        elif value > second and value != first:
            second = value

    return second

print(second_largest([10, 20, 4, 45, 99]))
```

```python
# 17. Write a function to check password strength using nested-if.
def password_strength(password):
    if len(password) >= 8:
        if "@" in password or "#" in password or "$" in password:
            return "Strong"
        else:
            return "Medium"
    else:
        return "Weak"

print(password_strength("abc@1234"))
```

```python
# 18. Write a function to return Fibonacci series using loops.
def fibonacci_series(n):
    a = 0
    b = 1
    result = []
    count = 0

    while count < n:
        result.append(a)
        c = a + b
        a = b
        b = c
        count = count + 1

    return result

print(fibonacci_series(8))
```

```python
# 21. Write a function to convert Instagram-like counts (e.g., 12000 -> "12K").
def format_instagram_count(count):
    if count >= 1000000:
        return str(count // 1000000) + "M"
    elif count >= 1000:
        return str(count // 1000) + "K"
    else:
        return str(count)

print(format_instagram_count(12000))
```

```python
# 50. Write a function to count "present" students in classroom attendance.
def count_present(attendance_list):
    count = 0
    for status in attendance_list:
        if status == "present":
            count = count + 1
    return count

print(count_present(["present", "absent", "present", "present"]))
```
