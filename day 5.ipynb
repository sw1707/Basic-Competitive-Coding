```python
# Question 1: Security key = count of repeating digits in given number.
# If no digit repeats, print -1.

def security_key_count(number):
    text = str(number)
    freq = {}

    for ch in text:
        if ch in freq:
            freq[ch] = freq[ch] + 1
        else:
            freq[ch] = 1

    repeating_count = 0
    for key in freq:
        if freq[key] > 1:
            repeating_count = repeating_count + 1

    if repeating_count == 0:
        return -1
    return repeating_count

# Example
print(security_key_count(12234))  # Output: 1
```

```python
# Question 2: Arrange list so all even numbers come first, then odd numbers.

def arrange_even_then_odd(values):
    evens = []
    odds = []

    for num in values:
        if num % 2 == 0:
            evens.append(num)
        else:
            odds.append(num)

    return evens + odds

# Example
nums = [5, 2, 7, 8, 1, 4]
result = arrange_even_then_odd(nums)
print(*result)
```

```python
# Question 3: Encrypt a character using key K (same case, cyclic for last letters).

def encrypt_character(ch, k):
    if 'A' <= ch <= 'Z':
        start = ord('A')
        pos = ord(ch) - start
        new_pos = (pos + k) % 26
        return chr(start + new_pos)
    elif 'a' <= ch <= 'z':
        start = ord('a')
        pos = ord(ch) - start
        new_pos = (pos + k) % 26
        return chr(start + new_pos)
    else:
        return ch

# Example
print(encrypt_character('D', 3))  # Output: G
```
